```
train_dir = 'train'
validation_dir = 'validate'
test_dir = 'test'
image_size = 200
```

```
#Training data moved to Extra folder to remove skewness
train_files=glob.glob(os.path.join(train_dir,"0","*.jpg"))
p = np.random.permutation(len(train_files))
for i in range(int(len(p)/2)):
    os.rename(train_files[i],"/content/Extra/"+train_files[i][8:])
```

```
train_files=glob.glob(os.path.join(train_dir,"*","*.jpg"))
validation_files=glob.glob(os.path.join(validation_dir,"*","*.jpg"))
test_files=glob.glob(os.path.join(test_dir,"*","*.jpg"))
```

```
from PIL import Image
train_images=[]
train_images_label=[]
test_images=[]
test_images_label=[]
validation_images=[]
validation_images_label=[]
for file in train_files:
    train_images_label.append(file[6])
    im=Image.open(file)
    im=im.resize((64,64))
    im=np.array(im)/255.
    im=im.flatten()
    train_images.append(im)
for file in validation_files:
    train_images_label.append(file[9])
    im=Image.open(file)
    im=im.resize((64,64))
    im=np.array(im)/255.
    im=im.flatten()
    train_images.append(im)
for file in test_files:
    test_images_label.append(file[5])
    im=Image.open(file)
    im=im.resize((64,64))
    im=np.array(im)/255.
    im=im.flatten()
    test_images.append(im)
```

```
X=np.array(train_images)
Y=np.array(train_images_label)
y=[]
for i in Y:
    if(i[0]=='0'):
        y.append([1.,0,0,0,0])
    if(i[0]=='1'):
        y.append([0,1.,0,0,0])
    if(i[0]=='2'):
        y.append([0,0,1.,0,0])
    if(i[0]=='3'):
        y.append([0,0,0,1.,0])
    if(i[0]=='4'):
        y.append([0,0,0,0,1.])
y=np.array(y)
```

```
# create model
model = Sequential()
model.add(Dense(1200, input_dim=12288, activation='relu'))
model.add(Dense(120, activation='relu'))
model.add(Dense(5, activation='softmax'))
```

```
model.summary()
```

```
from keras.callbacks import EarlyStopping
callbacks = [EarlyStopping(monitor='loss', patience=2, verbose=1, mode='min')]
```

```
from keras.optimizers import adam
opt = adam(lr=1e-3)
model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
```

```
model.fit(X, y, epochs=50, batch_size=32,callbacks=callbacks,shuffle=True)
```

```
X=np.array(test_images)
Y=np.array(test_images_label)
```

```
predictions = model.predict(X)
```

```
X=np.array(test_images)
Y=np.array(test_images_label)
y=[]
for i in Y:
    if(i[0]=='0'):
        y.append([1,0,0,0,0])
    if(i[0]=='1'):
        y.append([0,1,0,0,0])
    if(i[0]=='2'):
        y.append([0,0,1,0,0])
    if(i[0]=='3'):
        y.append([0,0,0,1,0])
    if(i[0]=='4'):
        y.append([0,0,0,0,1])
y=np.array(y)
```

```
from sklearn.metrics import classification_report
pred=np.argmax(predictions,axis=1)
y_true=np.argmax(y,axis=1)
print(classification_report(y_true,pred))
```
